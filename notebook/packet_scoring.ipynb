```python
from sklearn import datasets
from sklearn import linear_model
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import r2_score as r2
```

```python
# Load clusters and labels

clustered = pd.read_csv('/content/Exercise 6/xy.csv')
labeled = pd.read_csv('/content/Exercise 6/clustered.csv')
```

```python
clustered.head()
```

```python
labeled.head()
```

```python
# Load X and Y

X = clustered.drop(columns=['Unnamed: 0', 'Packet_Number', 'Timestamp', 'Length', 'Flags', 'Sequence_Number', 'x_pos', 'y_pos'])
y = labeled.cluster
X.head()
```

```python
y.head()
```

```python
# Train test split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)
```

```python
# Model parameters

model = linear_model.LogisticRegression(
    random_state = 0,
    solver = 'liblinear',
    #multi_class = 'auto',
    #n_jobs = 100
)
```

```python
# fit model

model.fit(X_train, y_train)
```

```python
# perform predictions

pred = model.predict(X_test)
prob = model.predict_proba(X_test)
```

```python
# Model Evaluation
"Train fit: ", model.score(X_train, y_train), "Test fit: ", model.score(X_test, y_test)
```

```python
# Predictions and actual
prob, pred, y_test
```

```python
# Coefficients
coeff = pd.DataFrame(model.coef_)
coeff.columns = X_train.columns
coeff
```

```python
# Intercept
inter = pd.DataFrame(model.intercept_)
inter
```
